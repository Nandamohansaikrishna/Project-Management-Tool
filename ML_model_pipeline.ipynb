```python
import torch

from transformers import AutoTokenizer, pipeline
```

```python
model="Qwen/Qwen2.5-3B-Instruct"
```

```python
tokenizer=AutoTokenizer.from_pretrained(model)
```

```python
from transformers import AutoModelForCausalLM, pipeline
import torch

# Explicitly load the model with the desired configurations
loaded_model = AutoModelForCausalLM.from_pretrained(
    model, # This 'model' refers to the string variable 'Qwen/Qwen2.5-3B-Instruct'
    torch_dtype=torch.bfloat16,
    device_map="auto",
    trust_remote_code=True
)

# Now, create the pipeline using the loaded model and tokenizer objects
pipeline = pipeline(
    "text-generation",
    model=loaded_model, # Pass the loaded model object
    tokenizer=tokenizer
)
```

```python
prompt="what is knn algorithm"
```

```python
newline_token=tokenizer.encode("\n")[0]

sequences=pipeline(
    prompt,
    eos_token_id=newline_token,
    pad_token_id=newline_token,
    do_sample=True,
    max_new_tokens=500, # Explicitly set a higher limit for new tokens
    return_full_text=False,
    num_return_sequences=1
)
print(sequences[0]['generated_text'])
```
